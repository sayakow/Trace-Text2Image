## region_bbox.h5

```python
# region_bbox.h5をコピーする

import shutil
shutil.copy('./data/'+data_name+'/region_bbox.h5', './data/'+data_name+"/"+str(n))
```

```python
f1 = h5py.File('./data/'+data_name+"/"+str(n)+"/region_bbox.h5", 'r+')
image = f1[data_name]
```

```python
from collections import namedtuple
Rectangle = namedtuple('Rectangle', 'xmin ymin xmax ymax')

def area(a, b):  # returns None if rectangles don't intersect
    dx = min(a.xmax, b.xmax) - max(a.xmin, b.xmin)
    dy = min(a.ymax, b.ymax) - max(a.ymin, b.ymin)
    if (dx>=0) and (dy>=0):
        return dx*dy
    else:
        return 0

ans = 0
remove_list = []

for box in range(len(image)):
    ra = Rectangle(x_min, y_min, x_max, y_max)
    rb = Rectangle(image[box][0], image[box][1], image[box][2], image[box][3])
    ans = area(ra, rb)
    print(ans)
    # 説明範囲を1/5含んでいなかったら
    if(ans < ((x_max-x_min)*(y_max-y_min))/5):
        #print("領域外なので削除済み -- ",image[box])
        image[box] = [0,0,0,0,0,0]
        # 領域外のindexを保存しておく
        remove_list.append(box)
    #重なっている部分が、bboxの1/3以下の場合
    if(ans < ((image[box][0]-image[box][2])*(image[box][1]-image[box][3]))/3):
        #print("領域外なので削除済み -- ",image[box])
        image[box] = [0,0,0,0,0,0]
        # 領域外のindexを保存しておく
        remove_list.append(box)
    print(image[box])
f1.close()
```
