```python
# クラス一覧
with open('./bbox_class.csv') as f:
    reader = csv.reader(f)
    bbox_class = [row for row in reader]
```

```python
print(df['annotations'][0][1]['image_id'])
#対象のオブジェクトを囲むバウンディングボックスの左上の頂点座標とサイズ（x, y, width, heightの順番）
print(df['annotations'][0][1]['bbox'])
print(df['annotations'][0][1]['category_id'])
```

```python
#データを探す
id_num = 34139

bbox_class_list_list = []
for i in range(len(df['annotations'][0])):
    bbox_class_list =[]
    if(df['annotations'][0][i]['image_id'] == id_num):
        bbox = df['annotations'][0][i]['bbox']
        category_id = df['annotations'][0][i]['category_id']
        #print(bbox,category_id)
        #print(bbox_class[category_id][1],bbox_class[category_id][2])
        bbox_class_list.append(bbox)
        bbox_class_list.append(bbox_class[category_id][2])
        bbox_class_list_list.append(bbox_class_list)
bbox_class_list_list
```

```python
#何秒にマウスがどこの位置にいるか
print("マウス位置 - 時間")

trace = data_1[0][5][0]

trace_list_list = []

for i in range(len(trace)):
    trace_list = []
    trace_list.append(trace[i]["x"])
    trace_list.append(trace[i]["y"])
    trace_list.append(trace[i]["t"])
    trace_list_list.append(trace_list)
trace_list_list
```

```python
for i in range(len(data_1[0][5][0])):
    data_1[0][5][0][i]["bbox"] = "None"
    x = data_1[0][5][0][i]["x"]*width
    y = data_1[0][5][0][i]["y"]*height
    #print(x,y)
    for j in range(len(bbox_class_list_list)):
        #print(x_min_bbox,y_min_bbox,x_max_bbox,y_max_bbox)
        x_min_bbox = bbox_class_list_list[j][0][0]
        y_min_bbox = bbox_class_list_list[j][0][1]
        x_max_bbox = x_min_bbox + bbox_class_list_list[j][0][2]
        y_max_bbox = y_min_bbox + bbox_class_list_list[j][0][3]
        #もしトレースの座標がbboxの中にあったら
        if((x < x_max_bbox)and(x_min_bbox < x)and(y < y_max_bbox)and(y_min_bbox < y)):
            #print(bbox_class_list_list[j][1])
            data_1[0][5][0][i]["bbox"] = bbox_class_list_list[j][1]
            #print(i,bbox_class_list_list[j][1])
```

```python
for i in range(len(trace_list_list)):
    x = trace_list_list[i][0]*width
    y = trace_list_list[i][1]*height
    for j in range(len(bbox_class_list_list)):
        #bboxの座標
        x_min_bbox = bbox_class_list_list[j][0][0]
        y_min_bbox = bbox_class_list_list[j][0][1]
        x_max_bbox = x_min_bbox + bbox_class_list_list[j][0][2]
        y_max_bbox = y_min_bbox + bbox_class_list_list[j][0][3]
        #もしトレースの座標がbboxの中にあったら
        if((x < x_max_bbox)and(x_min_bbox < x)and(y < y_max_bbox)and(y_min_bbox < y)):
            trace_list_list[i].append(bbox_class_list_list[j][1])
            #print(i,bbox_class_list_list[j][1])
```
